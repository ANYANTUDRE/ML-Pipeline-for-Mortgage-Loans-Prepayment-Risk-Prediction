# Data preparation ‚öôÔ∏è
We'll transform some columns as follow:  
**LTV                  --> LTV_range:**    Low 0-25, Medium 25-50, High 50-1000   
**CreditScore          --> Credit_range:** poor 0-500, fair 500-650, good 650-750, excellent 750-900  
**MonthsInRepayment    --> Repay_range:**  1: 0-4, 2: 4-8, 3: 8-12, 4: 12-16, 5: 16-20   
**FirstTimeHomebuyer   --> IsFirstTimeHomebuyer:** Yes/No 

```python
# LTV --> LTV_range
LTV_bins = [0, 25, 50, 1000]
LTV_categories = ['Low', 'Medium', 'High']

data['LTV_range'] = pd.cut(data['LTV'], bins=LTV_bins, labels=LTV_categories, right=False)


# CreditScore --> Credit_range
Credit_bins = [0, 500, 650, 750, 900]
Credit_categories = ['Poor', 'Fair', 'Good', 'Excellent']

data['Credit_range'] = pd.cut(data['CreditScore'], bins=Credit_bins, labels=Credit_categories, right=False)


# MonthsInRepayment  --> Repay_range
Repay_bins = [0, 4, 8, 12, 16, 20]
Repay_categories = ['1', '2', '3', '4', '5']

data['YearsInRepayment'] = data['MonthsInRepayment']/12
data['Repay_range'] = pd.cut(data['YearsInRepayment'], bins=Repay_bins, labels=Repay_categories, right=False)


# FirstTimeHomebuyer   --> IsFirstTimeHomebuyer
data['IsFirstTimeHomebuyer'] = data['FirstTimeHomebuyer'].map({  0: 'No',
                                                                 1: 'Yes'
                                                               })

data.to_csv(Prepared_Dataset.csv)
```

# Sample a test set (no data snooping!) üôà

Before exploring the data, we must sample a test set, put it aside, and never look at it.

**Why split the data at this stage?** 
Your brain is an amazing pattern detection system, which means that it is highly prone to overfitting: if you look at the test set, you may stumble upon some seemingly interesting pattern in the test data that leads you to select a particular kind of
Machine Learning model. When you estimate the generalization error using the test set, your estimate will be too optimistic and you will launch a system that will not perform as well as expected. This is called data snooping bias.

To create the test set we'll just pick some instances randomly, typically 10% of the dataset, and set them aside:

```python
train_set, test_set = train_test_split(data, test_size=0.1, random_state=42)
```

Instead of using pandas describe() method which doesn't offer some advanced informations, we will define our own function for univariate statistics with more informations such as the mode, skewness, kurtosis...

```python
# fonction to calculate univariate stats
def univariate_stats(df):
    output_df = pd.DataFrame(columns=['Count', 'Missing', 'Unique', 'Dtype', 'IsNumeric', 'Mode', 'Mean', 'Min', '25%', 'Median', '75%', 'Max', 'Std', 'Skew', 'Kurt'])
    
    for col in df:
        if is_numeric_dtype(df[col]):
            output_df.loc[col] = [df[col].count(), df[col].isnull().sum(), df[col].nunique(), df[col].dtype, is_numeric_dtype(df[col]), df[col].mode().values[0], df[col].mean(), df[col].min(), df[col].quantile(.25), df[col].median(), df[col].quantile(.75), df[col].max(), df[col].std(), df[col].skew(), df[col].kurt() ]
        else:
            output_df.loc[col] = [df[col].count(), df[col].isnull().sum(), df[col].nunique(), df[col].dtype, is_numeric_dtype(df[col]), df[col].mode().values[0], '-', '-', '-', '-', '-', '-', '-', '-', '-' ]

    return output_df.sort_values(by=['IsNumeric', 'Unique'], ascending=False)


pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)


# Call the function to check univariate stats on the dataset
univariate_stats(train_set)
```

# 2Ô∏è‚É£ | Bivariate Analysis üìä

Recall üí°: There are three types of bivariate analysis.       

- **Numeric - Numeric:** Pearson's Correlation  
The correlation represents the strength of a linear relationship between two numerical variables. If there is no correlation between the two variables, there is no tendency to change along with the values of the second quantity.
- **Categoric - Numeric:** one-way ANOVA(3 + groups) or t-test (exactly 2 groups)  
The ANOVA test is used to determine whether there is a significant difference among the averages of more than two groups that are statistically different from each other.
- **Categoric - Categoric:** Chi-square Test for example  
It is calculated based on the difference between expected frequencies and the observed frequencies in one or more categories of the frequency table.

```python
# this just an intermediate function that will be used in bivstats for one-way ANOVA
def anova(df, feature, label):
    import pandas as pd
    import numpy as np
    from scipy import stats
    
    groups = df[feature].unique()
    df_grouped = df.groupby(feature)
    group_labels = []
    for g in groups:
        g_list = df_grouped.get_group(g)
        group_labels.append(g_list[label])
        
    return stats.f_oneway(*group_labels)

# function to calculate bivariate stats; Pearson' correlation, p-value and one-way ANOVA
def bivstats(df, label):
    from scipy import stats
    import pandas as pd
    import numpy as np
    
    # Create an empty DataFrame to store output
    output_df = pd.DataFrame(columns=['Type of bivariate analysis', 'Stat Name', 'Stat symbol', '+1/-1 * ', 'Effect size', 'p-value'])
    
    for col in df:
        if col != label:
            if df[col].isnull().sum() == 0:
                if is_numeric_dtype(df[col]):   # Only calculate r, 
                    r, p = stats.pearsonr(df[label], df[col])
                    output_df.loc[col] = ['Numeric - Numeric', 'Pearson', 'r', np.sign(r), abs(round(r, 3)), round(p,8)]
                    
                else:
                    F, p = anova(df[[col, label]], col, label)
                    output_df.loc[col] = ['Categoric - Numeric', 'ANOVA', 'F', '', round(F, 3), round(p,8)]
                    
            else:
                output_df.loc[col] = [np.nan, np.nan, np.nan, np.nan, np.nan, np.nan]

    return output_df.sort_values(by=['Effect size', 'Stat symbol'], ascending=[False, False])
 
pd.options.display.float_format = '{:.5f}'.format
bivstats(train_set, target)
```

**Insights && Observations üí° :**
- From ANOVA test results(F) above we can say that categorical variables Credit_range, Repay_range and LTV_range are mostly correlated to the target variable EverDelinquent: they are the three(03) most important columns.
- Numerical variables: MonthsDelinquent, NumBorrowers, OrigUPB are mostly correlated( Pearson's r) to the target variable EverDelinquent.

```python
def plot_heatmap(df, title):
    # Create a mask for the diagonal elements
    mask = np.zeros_like(df.astype(float).corr())
    mask[np.triu_indices_from(mask)] = True

    # Set the colormap and figure size
    colormap = plt.cm.RdBu_r
    plt.figure(figsize=(16, 16))

    # Set the title and font properties
    plt.title(f'{title} correlation of features', fontweight='bold', y=1.02, size=24)

    # Plot the heatmap with the masked diagonal elements
    sns.heatmap(df.astype(float).corr(), linewidths=0.1, vmax=1.0, vmin=-1.0, 
                square=True, cmap=colormap, linecolor='white', annot=True, annot_kws={"size": 14, "weight": "bold"},
                mask=mask)

plot_heatmap(train_set[numeric_cols+binary_cols+[target]], title='Data')
```

**Insights && Observations üí° :**

## Clustering Analysis  
Clustering is part of an unsupervised algorithm in Machine Learning, meaning no training process is involved. Instead, organizing data into groups with related characteristics and categorizing them is the goal of clustering.   
THis data mining technique allows us to understand the data distribution based on available features.  
Let's implement K-means clustering algorithm over our dataset.

### 1. Feature Scaling

```python
# Before we conduct the scaling, let‚Äôs create a new data frame consisting of the selected column for the scaling process.
selected_columns_for_Kmeans = numeric_cols + binary_cols
X = train_set[selected_columns_for_Kmeans]
```

```python
# Standardize the numerical variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

###  2. Number of Clusters  
Determining the optimal number of clusters into which data can be clustered is a fundamental step for any unsupervised algorithm. **The Elbow Method** is one of the most widely used approaches for determining the optimal value of ‚ÄúK‚Äù.

```python
# Determine the optimal number of clusters using the elbow method
inertia = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.xticks(range(1, 11))
plt.show()
```

**As I'm hesitating to determine the K value from the elbow method graph, I'll use a help of a function called KneeLocater from the kneed library.**

## Principal Component Analysis - PCA 
PCA is a dimentionality reduction technique frequently used to reduce the dimensions of large or big datasets.

```python
# Let's first combine categorical and binary data into a single DataFrame
categorical_data = pd.get_dummies(train_set[categ_cols])
binary_data = train_set[binary_cols]
combined_data = pd.concat([categorical_data, binary_data], axis=1)
```

```python
# Standardize numerical data
scaler = StandardScaler()
numeric_data = scaler.fit_transform(train_set[numeric_cols])
```

```python
# now we'll perform PCA on the combined data
pca = PCA()
pca_result = pca.fit_transform(np.concatenate((numeric_data, combined_data), axis=1))

print(pca.explained_variance_ratio_ * 100)
```

```python
# Explained variance ratio plot
plt.figure(figsize=(10, 6))
plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Explained Variance by Number of Components')
plt.show()
```

```python
# Choose the number of components based on explained variance: a suitable threshold (e.g., 95% explained variance)
num_components = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.95) + 1
num_components
```

```python
# Perform PCA again with the chosen number of components
pca_final = PCA(n_components=num_components)
pca_result_final = pca_final.fit_transform(np.concatenate((numeric_data, combined_data), axis=1))

# Visualize the PCA result (for example, the first two components)
plt.figure(figsize=(10, 6))
plt.scatter(pca_result_final[:, 0], pca_result_final[:, 1], c=train_set['Cluster'], cmap='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA Result')
plt.colorbar(label='Cluster')
plt.show()
```
