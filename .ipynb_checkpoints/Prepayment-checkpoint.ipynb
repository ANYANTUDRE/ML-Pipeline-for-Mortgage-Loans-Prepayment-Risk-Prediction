#### Removing Outliers

- It seems there are a few features with outliers.
- Therefore, Using the below function we will remove any rows with n or more outliers in a single row.

```python
from collections import Counter

def detectOutliers(data, n, features):
    outlier_rows = []
    for col in features:
        Q1 = np.percentile(data[col], 25)
        Q3 = np.percentile(data[col], 75)
        IQR = Q3 - Q1
        outlierStep = IQR * 1.5
        outlierList = data[(data[col] < Q1 - outlierStep ) | (data[col] > Q3 + outlierStep)].index
        outlier_rows.extend(outlierList)
    # Counts the frequency of occurrences of each row index
    outlier_rows = Counter(outlier_rows)
    multiple_outliers = list(key for key, value in outlier_rows.items() if value >= n)
    return multiple_outliers
```

```python
numeric_columns = ["MSA", "MIP", "Units", "OCLTV", "DTI", "OrigUPB", "LTV", "OrigInterestRate", "OrigLoanTerm", "NumBorrowers"]
outliers_to_drop = detectOutliers(data, 2, numeric_columns)
print("these rows will be dropped due to their outlier values: {} ".format(outliers_to_drop))
len(outliers_to_drop)
```

```python
print("Before removing the outliers we have {} rows".format(len(data)))
data = data.drop(outliers_to_drop, axis=0).reset_index(drop=True)
print("After removing the outliers we have {} rows".format(len(data)))
```

```python
def plot_histogram(data, feature):
  """
  Plots a histogram of the specified feature in the given data.

  Args:
    data: The data to plot the histogram for.
    feature: The name of the feature to plot.

  Returns:
    A matplotlib Axes object containing the plot.
  """
  # Create a new figure and a subplot for the histogram
  fig, ax = plt.subplots()

  # Plot the histogram using seaborn
  sns.histplot(data=data, x=feature, ax=ax, kde=True, hue="EverDelinquent")

  # Customize the plot
  ax.set_title(f"Histogram of {feature}")
  ax.set_xlabel(feature)
  ax.set_ylabel("Frequency")

  return ax
```

### Prepayment

#### Equated Monthly Installment - EMI

Equated Monthly Installment - EMI for short - is the amount payable every month to the bank or any other financial institution until the loan amount is fully paid off. It consists of the interest on loan as well as part of the principal amount to be repaid.

##### For EMI

```python
monthlyInterestRate = data["OrigInterestRate"]/(12 * 100)
data["monthly_payment"] = data["OrigUPB"] * monthlyInterestRate * (((1 + monthlyInterestRate)**data["OrigLoanTerm"])/(((1 + monthlyInterestRate)**data["OrigLoanTerm"]) - 1))
```

##### For total payment

```python
data["totalPayment"] = data["monthly_payment"] * data["OrigLoanTerm"]
data["interestAmount"] = data["totalPayment"] - data["OrigUPB"]
```

##### Calculating Monthly Income

- DTI is the ration of debt to income
- Therefore, monthly income = EMI / DTI
- Which means total income = monthly income * Original Loan Term

```python
data["monthly_income"] = (data["totalPayment"] /data["DTI"])
```

##### Calculate current principal

```python
def principal(rate, amount, emi, month):
  for i in range(month):
    interest = rate * amount
    p = emi - interest
    amount -= p
  return amount

data["currentPrincipal"] = np.vectorize(principal)(monthlyInterestRate, data["OrigUPB"], data["monthly_payment"], data["MonthsInRepayment"])
```

##### Calculating Pre-payment

For having a simple model, we are assuming that a person is going to save up money based on their DTI for 2 years and pre-pay the residual amount after subtracting it with the EMI of those 2 years. We have also assumed that a person with DTI less than 40 is going to save up 50% of his income while a person with DTI more than 40% is going to save up 75% of his income and these savings are inclusive of EMI, so for example if a person has DTI of 25% and he is going to pre-pay 50-25=25% of his income.

```python
def compute_pre_payment(dti, income):
  if dti < 40:
    return income/2
  else:
    return income * (3/4)

data["prePayment"] = np.vectorize(compute_pre_payment)(data["DTI"], data["monthly_income"] * 24)
data["prePayment"] = data["prePayment"] - data["monthly_payment"]*24
```

```python
data["MonthsOfPrepayments"] = data["prePayment"]/data["monthly_payment"]
```

```python
data["scheduledPayments"] = data["monthly_payment"] * (data["MonthsInRepayment"] + data["MonthsOfPrepayments"] + data["MonthsDelinquent"])
```

```python
data[data["prePayment"] < 0]
```

```python
data["prepayment_risk"] = (data["scheduledPayments"] - (data["monthly_payment"] * data["MonthsInRepayment"]))/data["currentPrincipal"]
```

### Mutual Information

```python
# Utility functions
def make_mi_scores(X, y):
    X = X.copy()
    for colname in X.select_dtypes(["object", "category"]):
        X[colname], _ = X[colname].factorize()
    # All discrete features should now have integer dtypes
    discrete_features = [pd.api.types.is_integer_dtype(t) for t in X.dtypes]
    mi_scores = mutual_info_regression(X, y, discrete_features=discrete_features, random_state=0)
    mi_scores = pd.Series(mi_scores, name="MI Scores", index=X.columns)
    mi_scores = mi_scores.sort_values(ascending=False)
    return mi_scores


def plot_mi_scores(scores):
    scores = scores.sort_values(ascending=True)
    width = np.arange(len(scores))
    ticks = list(scores.index)
    plt.barh(width, scores)
    plt.yticks(width, ticks)
    plt.title("Mutual Information Scores")
```

```python
X = data.copy()
y = X.pop("prepayment_risk")

mi_scores = make_mi_scores(X, y)
```
